```python
class Motor:
  def __init__(self, power_kw, efficiency):
    self.power_kw = power_kw
    self.efficiency = efficiency

  def output_kw(self):
    return self.power_kw * self.efficiency

motor1 = Motor(200, 0.8)
print(motor1.output_kw())
```

Here is an example of basic exception handling using `try` and `except`:

```python
try:
  # This code might raise a ZeroDivisionError
  result = 10 / 0
  print(result)
except ZeroDivisionError:
  print("Error: Cannot divide by zero!")
```

You can also catch multiple types of exceptions:

```python
try:
  # This code might raise a ValueError or a TypeError
  num = int("abc")
  print(num)
except (ValueError, TypeError):
  print("Error: Invalid value or type!")
```

Using `else` and `finally` with `try` and `except`:

```python
try:
  # This code will not raise an exception
  result = 10 / 2
except ZeroDivisionError:
  print("Error: Cannot divide by zero!")
else:
  print("Division successful, result:", result)
finally:
  print("This finally block is always executed.")
```

```python
import requests

# Define the API endpoint
api_url = "https://jsonplaceholder.typicode.com/users"

try:
    # Send a GET request to the API
    response = requests.get(api_url)

    # Check if the request was successful (status code 200)
    response.raise_for_status()

    # Parse the JSON response
    users_data = response.json()

    # Print the data
    print("Successfully fetched user data:")
    for user in users_data:
        print(f"- Name: {user['name']}, Email: {user['email']}")

except requests.exceptions.RequestException as e:
    print(f"Error fetching data from the API: {e}")
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private variable

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def get_balance(self):
        return self.__balance

# Usage
acc = BankAccount(1000)
acc.deposit(500)
print("Balance:", acc.get_balance())
```

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

# Usage
a = Animal()
a.speak()

d = Dog()
d.speak()
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car engine started")

# Usage
c = Car()
c.start()
```
